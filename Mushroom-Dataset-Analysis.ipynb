Data Manipulation

```python
data = data.astype('category')
```

```python
data.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in data.columns:
    data[column]=le.fit_transform(data[column])
```

```python
data.head()
```

Store feature matrix in X and Response(Target) in vector y

```python
X = data.drop('class', axis=1)
y = data['class']
```

```python
X
```

Applying PCA
#using to convert large dataset into small which contains most of the information

```python
from sklearn.decomposition import PCA
pca1 = PCA(n_components=7)
pca_fit = pca1.fit_transform(X)
```

```python
pca1.explained_variance_ratio_
```

```python
sum(pca1.explained_variance_ratio_)
```

Splitting the dataset into Training and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(pca_fit,y,test_size=0.20,random_state=42)
```

```python
X_train
```

```python
y_train
```

Import the models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

Model Training

```python
import numpy as np
lr = LogisticRegression()
lr.fit(X_train,y_train)

knn = KNeighborsClassifier()
knn.fit(X_train,y_train)

svc = SVC()
svc.fit(X_train,y_train)

dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)

rm = RandomForestClassifier()
rm.fit(X_train,y_train)

gb = GradientBoostingClassifier()
gb.fit(X_train,y_train)
```

Prediction on Test Data

```python
y_pred1 = lr.predict(X_test)
y_pred2 = knn.predict(X_test)
y_pred3 = svc.predict(X_test)
y_pred4 = dt.predict(X_test)
y_pred5 = rm.predict(X_test)
y_pred6 = gb.predict(X_test)
```

Evaluating the alogirthm

```python
from sklearn.metrics import accuracy_score
```

```python
print("ACC LR",accuracy_score(y_test,y_pred1))
print("ACC KNN",accuracy_score(y_test,y_pred2))
print("ACC SVC",accuracy_score(y_test,y_pred3))
print("ACC DT",accuracy_score(y_test,y_pred4))
print("ACC RM",accuracy_score(y_test,y_pred5))
print("ACC GBC",accuracy_score(y_test,y_pred6))
```

```python
final_data = pd.DataFrame({'Models':['LR','KNN','SVC','DT','RM','GBC'], 'ACC': [accuracy_score(y_test,y_pred1)*100,accuracy_score(y_test,y_pred2)*100,accuracy_score(y_test,y_pred3)*100,accuracy_score(y_test,y_pred4)*100,accuracy_score(y_test,y_pred5)*100,accuracy_score(y_test,y_pred6)*100]})
```
